### Задача 1

Создайте датафрейм, сохранив его в отдельном Excel файле Output data.xlsx, со следующими свойствами:
- Столбец month должен быть преобразован в столбец days, содержащий абсолютно точное количество дней, которое человек проживает по текущему адресу. При преобразовании предполагайте, что сегодня 2000-01-01.
- Добавьте столбец с доходом на одного члена семьи (income_per_family_member). При расчете четко опишите в комментарии, какие предположения вы сделали для определения количества членов семьи.
- Выберите свой собственный критерий определения уровня дохода, разбейте данные согласно выбранному критерию, записав результат в новый столбец income_level. Подробно опишите выбранный критерий.
- Добавьте столбец, содержащий отношение средних годовых трат по кредитной карте к годовому доходу (share_of_annual_expentitures).
- Итоговый датафрейм должен содержать информацию только о людях, для которых share_of_annual_expenditures превышает 100%.
- Отсортируйте итоговый датафрейм таким образом, чтобы первые строки содержали информацию о самозанятых людях. Далее, отдельно для самозанятых и не самозанятых, отсортируйте датафрейм в соответствии с share_of_annual_expenditures (от большего к меньшему).

```python
#Импортируем оригинальную таблицу
import pandas as pd
frame = pd.read_csv('data_assignment2.csv', header=0, sep = '/')
frame
```

```python
#считаем абсолютно точное число дней с заезда на текущее жилье до 2000-01-01
import datetime

#создаем и заполняем list now текущей датой
now = [datetime.date(2000, 1, 1) for i in range(0, len(frame.months))]

#создаем два списка с годом и месяцем (в соотвествии с кол-во месяцев проживания)
then_year = 1999-frame.months//12
then_month = 13-frame.months%12

#если число месяцев делится на 12, frame.months%12 = 0 => исправляем месяц
for i in range(0, len(frame.months)):
    if then_month[i] == 13:
        then_month[i]=1
        
#создаем и заполняем список then с датами заселения по текущему адресу
then = [0 for i in range(0, len(frame.months))]
for i in range(0, len(frame.months)):
    then[i] = datetime.date(then_year[i], then_month[i], 1)
    
#создаем и заполняем список then с числом дней с заселения
days_temp = [0 for i in range(0, len(frame.months))]
for i in range(0, len(frame.months)):
    days_temp[i] = now[i] - then[i]
    
#переводим из формата timedelta в int  
days = [days_temp[i].days for i in range(0, len(frame.months))]

frame_output = frame
frame_output = frame_output.rename(columns={'months': 'days'})
frame_output['days'] = days

#т.к. мы не можем точно определить, работает ли жена(муж) владельца карты (входит ли в dependents),
#будем считать число членов семьи как dependents + 1

frame_output['income_per_family_member'] = frame_output.income / (frame_output.dependents + 1)

#будем разбивать на 5 уровней дохода - до 20 тыс.$, 20-40 тыс.$, 40-60 тыс.$, 60-80 тыс.$, свыше 80 тыс.$.
frame_output['income_level'] = 0

frame_output.loc[frame_output.income<2, 'income_level'] = 1
frame_output.loc[(frame.income>=2) & (frame.income<4), 'income_level'] = 2
frame_output.loc[(frame.income>=4) & (frame.income<6), 'income_level'] = 3
frame_output.loc[(frame.income>=6) & (frame.income<8), 'income_level'] = 4
frame_output.loc[frame.income>8, 'income_level'] = 5

#ВАЖНО!!! в описании ошибка - столбец share - это ежежегодные!!! траты на карту к доходу, т.е.
#expenditure = income * share / 12
#будем считать, что share - ежемесячные траты к ежегодным доходам
frame_output['share_of_annual_expentitures'] = frame_output.share * 12

#датафрейм содержит только тех, у гого share_of_annual_expentitures > 100% (т.е. 1)
frame_output=frame_output[frame_output.share_of_annual_expentitures > 1]

#сортировка списка по share_of_annual_expentitures (сначала идут самозанятые)
frame_output1 = frame_output[frame_output.selfemp=='yes']
frame_output2 = frame_output[frame_output.selfemp=='no']
frame_output1 = frame_output1.sort_values('share_of_annual_expentitures', ascending=False)
frame_output2 = frame_output2.sort_values('share_of_annual_expentitures', ascending=False)

frame_output=frame_output1
frame_output=frame_output.append(frame_output2)

#записываем в файл
writer = pd.ExcelWriter('Output data.xlsx', engine='xlsxwriter')
frame_output.to_excel(writer, sheet_name='OutputData', encoding='utf-8')
writer.save()
frame_output
```

- Средний возраст для каждого уровня доходов согласно выбранной вами классификации.
- Топ 5 самых молодых и самых пожилых людей, чью заявку на кредитную карту не удовлетворили.

```python
#Среднее количество dependents для людей, которые не являются собственниками жилья.
import statistics
print ('Среднее кол-во dependents для не собственников жилья: '
+str(statistics.mean(frame[frame.owner == 'no'].dependents)))
```

```python
#Средний возраст для каждого уровня доходов согласно выбранной вами классификации.
#будем разбивать на 5 уровней дохода - до 20 тыс.$, 20-40 тыс.$, 40-60 тыс.$, 60-80 тыс.$, свыше 80 тыс.$.
import statistics
print ('Средний возраст для уровня доходов 1 (до 20 тыс. $): ' 
       + str(statistics.mean(frame[(frame.income<2)].age)))
print ('Средний возраст для уровня доходов 2 (от 20 до 40 тыс. $): '
       + str(statistics.mean(frame[(frame.income>=2) & (frame.income<4)].age)))
print ('Средний возраст для уровня доходов 3 (от 40 до 60 тыс. $): ' 
       + str(statistics.mean(frame[(frame.income>=4) & (frame.income<6)].age)))
print ('Средний возраст для уровня доходов 4 (от 60 до 80 тыс. $): ' 
       + str(statistics.mean(frame[(frame.income>=6) & (frame.income<8)].age)))
print ('Средний возраст для уровня доходов 5 (свыше 80 тыс. $): ' 
       + str(statistics.mean(frame[frame.income>=8].age)))
```

```python
#Топ 5 самых молодых людей, чью заявку на кредитную карту не удовлетворили.
print ('Топ-5 самых молодых, чью заявку не одобрили: ')
frame[frame.card=='no'].sort_values('age').head(5)
```

```python
#Топ 5 самых пожилых людей, чью заявку на кредитную карту не удовлетворили.
print ('Топ-5 самых пожилых, чью заявку не одобрили: ')
frame[frame.card=='no'].sort_values('age',ascending=False).head(5)
```

```python
#Среднее количество major credit cards для 10 людей, обладающих самым высоким доходом.
import statistics
print('Среднее количество major credit cards у топ-10 с самым высоким доходом:')
statistics.mean(frame.sort_values('income',ascending=False).head(10).majorcards)
```
